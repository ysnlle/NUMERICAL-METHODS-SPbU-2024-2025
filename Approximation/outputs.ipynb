```python
import matplotlib.pyplot as plt
import help
import numpy as np
import methods
from prettytable import PrettyTable
```

# Normal Equations

```python
n = 5
xn = help.randNodes(60)
fn = [help.f(x) for x in xn]
xs = np.linspace(-1, 1, 100)
fs = [help.f(x) for x in xs] 
sample, xn = help.generator(xn)
plt.figure(figsize = (15,25))
tableNorm = []
for i in range(1, n+1):
    coefs = methods.normEqCoefs(xn, sample, i)
    approx = methods.normEq(xs, coefs)
    ax = plt.subplot(n, 1, i)
    ax.plot(xs, fs, label=" ", linewidth=1, color='blue')
    ax.plot(xs, approx, label=f"   {i}", linewidth=1, color='red')
    ax.scatter(xn, sample, color='black', s=10, label="")
    ax.legend()
    ax.grid()
    ax.set_title(f"  {i}")
    tableNorm.append(help.loss(fs, approx))
plt.tight_layout()  
plt.show()
```

# Ortogonal Polynomials


```python
n = 5
plt.figure(figsize = (15,25))
tableOrt = []
for i in range(1, n+1):
    approx = methods.approx(xs, xn, sample, i)
    ax = plt.subplot(n, 1, i)
    ax.plot(xs, fs, label=" ", linewidth=1, color='blue')
    ax.plot(xs, approx, label=f"   {i}", linewidth=1, color='red')
    ax.scatter(xn, sample, color='black', s=10, label="")
    ax.legend()
    ax.grid()
    ax.set_title(f"  {i}")
    tableOrt.append(help.loss(fs, approx))
plt.tight_layout()  
plt.show()
```

# Loss Function Values


```python
table = PrettyTable()
table.field_names = ['Polynomial Degree', 'Normal Equations', 'Ortogonal Polynomials']
for i in range(n):
    table.add_row([i+1, tableNorm[i], tableOrt[i]])
table
```
